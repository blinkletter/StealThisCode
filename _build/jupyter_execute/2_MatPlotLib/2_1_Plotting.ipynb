# Plotting Data

Steal the code below to **plot data** in a **simple x,y plot**. Note that we set x and y to be two different lists of values. You can set the x and y to any two sets of data that have the same number of entries. One list might be rates and another might be concentrations of reactants, for example. Consider the code below.

````python
### Set up the data lists

conc = [1.1, 2.0, 2.8, 3.5, 5.0, 7.5, 9.8]     ### units are mM
rates = [18.7, 28.6, 35.0,  39.4, 46.5, 54.1, 58.5]   ### units are uM/min
````

````python
from matplotlib import pyplot as plt       ### load in the plotting tools from the matplotlib package

### set x and y so that we never need to change the plotting code below

x = conc
y = rates

### A basic plot of x,y data

plt.plot(x,y)
````

## Settings and Features

the *PyPlot* sublibrary of the *MatPlotLib* package has many commands that will enable us to **control the style and feature**s of our plot. Consider the code below.

````python
### A slightly better plot. Take note of how these commands change the plot.

### Create blank plot
plt.rcdefaults()             ### resets the plot defaults so we always start in the same place
plt.figure(figsize=(4,3))    ### sets the size of the plot figure in inches

### Plot the data
plt.plot(x, y, "ko")         ### Plot the x and y data as black "k" circles "o"

### All the rest of these commands are about style
plt.xlim(0, None)            ### x-axis goes from 0 to limit of data
plt.ylim(0, None)            ### y-axis goes from 0 to limit of data

plt.title("Michaelis-Menten Plot")   ### use a title
plt.xlabel(r"$[S]$")                 ### label the x-axis. r means "raw" text string. $-signs are commands for the Latex math language 
plt.ylabel(r"$\nu$")                 ### label y-axis. Here we use the greek letter nu for rate.

### Output the plot
plt.savefig("plot.pdf")     ### export the plot as this
plt.show()                  ### display the plot in this notebook
````

## More Examples of Formatting

The *PyPlot* sublibrary has access to many many **features and controls** for plotting. Consider the example below.

````python
### More style examples

### Create blank plot
plt.rcdefaults()             ### resets the plot defaults so we always start in the same place
plt.figure(figsize=(4,3))    ### sets the size of the plot figure in inches

### Plot the data with style
plt.plot(x, y, 
        marker='o',                   ### Plot the x and y data 
        linestyle = 'None',           ### no line between markers
        markerfacecolor='lightgray',  ### markers and lines are this color
        markeredgecolor = 'black',    ### outline of markers is this color
        markeredgewidth = 0.5,        ### outline of markers is this wide
        markersize=8)                 ### markers are this big

### Add more style
plt.xlim(0, None)          ### x-axis goes from 0 to limit of data
plt.ylim(0, 70)            ### y-axis goes from 0 to 70

plt.title("Michaelis-Menten Plot")   ### use a title
plt.xlabel(r"$[S]$")                 ### label the x-axis. r means "raw" text string. $-signs are commands for the Latex math language 
plt.ylabel(r"$\nu$")                 ### label y-axis. Here we use the greek letter nu for rate.

plt.gca().spines['right'].set_visible(False)   ### Make right spine invisible
plt.gca().spines['top'].set_visible(False)     ### Make top spine invisible
plt.gca().spines['left'].set_visible(False)    ### Make top spine invisible
plt.gca().spines['bottom'].set_visible(False)  ### Make top spine invisible

### Output the plot
plt.savefig("plot.pdf")     ### export the plot as this
plt.show()                  ### display the plot in this notebook
````

## Plotting Functions
We will **plot** an exponential decay curve. We will plot *y = e<sup>–x</sup>*. First we need a ***x*-axis**. As you see below, we will use the `linspace` tool from *NumPy* to create an evenly spaced array of values between two points. We will then calculate a new array for the ***y*-axis** by applying the exponential decay function.  We now have a set of x,y values to plot.

We will then **plot the *x,y* data** using tools from the `plt` object (which is actually the ```matplotlib.pyplot``` library.) Now we will have our plot.

````python
import numpy as np

x = np.linspace(0,5,100)  # make an array of 100 evenly paced values beltween 0 and 5.
y = np.exp(-x)            # make an array of calculated values

plt.plot(x,y)             # plot y vs. x
plt.show()                # show the plot
````

## The Same Thing, Only Better
We can add elements to the plot such as **titles and labels**. We can set limits on the *x*- and *y*-axis. We can **change the color** of the line. Examine the code below and see how we did all that.

The **commands** executed below are called "methods".  They are functions built into the `plt` **object**. They modify the object according to the input for the method. When you enter `plt.ylabel("Amount")`, the `plt` object has a **label** added to its *y*-axis. In the last line, the `plt.show()` method **displays and clears** the `plt` object.  Entering `plt.show()` a second time will display nothing unless you build a new plot. 

There are much **more sophisticated** ways to present plots and we will explore some aspects of this later. However, this is all the **style** that we need to know for now.

````python
import numpy as np                 # A library of math functions and data structure
import matplotlib.pyplot as plt    # A library of plotting functions

x = np.linspace(0,5,100)           # make an array of 100 evenly paced values beltween 0 and 5.
y = np.exp(-x)                     # make an array of calculated values

plt.plot(x,y, "k--")               # plot y vs. x: the "k" means black, "--" means dashed line

plt.title("Exponential Decay")     # Title of the plot         
plt.xlabel("Time")                 # x-axis label   
plt.ylabel("Amount")               # y-axis label  

plt.xlim([-1,5])                   # set x-axis range
plt.ylim([0,1.2])                  # set y-axis range

plt.show()                         # show the plot
````

## Two Plots in a Figure

Let us **demonstrate** first order **reaction kinetics**. We will plot the concentration of reactant and product for the reaction *A &rarr; B*. We know that the instantaneous rate at any given time point is&hellip;

$$-\frac{\partial [A]}{\partial t}=\frac{\partial [B]}{\partial t}=k[A] $$ 

&hellip;and the integrated rate equation that describes the concentrations of *A* and *B* as they change over time are 

$$ [A]_t=[A]_0 e^{-k t}  \text{ and }  [B]_t=[A]_0 \left( 1-e^{-k t} \right)$$ 

Below we will create an array of values for the *x*-axis and then make two calculated arrays for the *y*-axis. One set is for [*A*] and the other is for [*B*]. We will then plot them.

````python
A0 = 1.5                  # create a variable that represents the initial concentration
k = 0.34                  # a variable that represents the rate constant

t = np.linspace(0,5,100)  # make an array of 100 evenly paced values beltween 0 and 5 for time
yA = A0*np.exp(-k*t)      # make an array of calculated values for [A]
yB = A0*(1-np.exp(-k*t))  # make an array of calculated values for [B]

plt.plot(x,yA)            # plot y vs. x
plt.plot(x,yB)            # plot y vs. x

plt.show()                # show the plot
````

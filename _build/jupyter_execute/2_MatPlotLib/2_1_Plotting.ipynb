# Plotting Data

Steal the code below to **plot data** in a **simple x,y plot**. Note that we set x and y to be two different lists of values. You can set the x and y to any two sets of data that have the same number of entries. One list might be rates and another might be concentrations of reactants, for example. Consider the code below.

```python
### Set up the data lists

conc = [1.1, 2.0, 2.8, 3.5, 5.0, 7.5, 9.8]     ### units are mM
rates = [18.7, 28.6, 35.0,  39.4, 46.5, 54.1, 58.5]   ### units are uM/min
```

```python
from matplotlib import pyplot as plt       ### load in the plotting tools from the matplotlib package

### set x and y so that we never need to change the plotting code below

x = conc
y = rates

### A basic plot of x,y data

plt.plot(x,y)
```

## Settings and Features

the *PyPlot* sublibrary of the *MatPlotLib* package has many commands that will enable us to **control the style and feature**s of our plot. Consider the code below.

```python
### A slightly better plot. Take note of how these commands change the plot.

### Create blank plot
plt.rcdefaults()             ### resets the plot defaults so we always start in the same place
plt.figure(figsize=(4,3))    ### sets the size of the plot figure in inches

### Plot the data
plt.plot(x, y, "ko")         ### Plot the x and y data as black "k" circles "o"

### All the rest of these commands are about style
plt.xlim(0, None)            ### x-axis goes from 0 to limit of data
plt.ylim(0, None)            ### y-axis goes from 0 to limit of data

plt.title("Michaelis-Menten Plot")   ### use a title
plt.xlabel(r"$[S]$")                 ### label the x-axis. r means "raw" text string. $-signs are commands for the Latex math language 
plt.ylabel(r"$\nu$")                 ### label y-axis. Here we use the greek letter nu for rate.

### Output the plot
plt.savefig("plot.pdf")     ### export the plot as this
plt.show()                  ### display the plot in this notebook
```

## More Examples of Formatting

The *PyPlot* sublibrary has access to many many **features and controls** for plotting. Consider the example below.

```python
### More style examples

### Create blank plot
plt.rcdefaults()             ### resets the plot defaults so we always start in the same place
plt.figure(figsize=(4,3))    ### sets the size of the plot figure in inches

### Plot the data with style
plt.plot(x, y, 
        marker='o',                   ### Plot the x and y data 
        linestyle = 'None',           ### no line between markers
        markerfacecolor='lightgray',  ### markers and lines are this color
        markeredgecolor = 'black',    ### outline of markers is this color
        markeredgewidth = 0.5,        ### outline of markers is this wide
        markersize=8)                 ### markers are this big

### Add more style
plt.xlim(0, None)          ### x-axis goes from 0 to limit of data
plt.ylim(0, 70)            ### y-axis goes from 0 to 70

plt.title("Michaelis-Menten Plot")   ### use a title
plt.xlabel(r"$[S]$")                 ### label the x-axis. r means "raw" text string. $-signs are commands for the Latex math language 
plt.ylabel(r"$\nu$")                 ### label y-axis. Here we use the greek letter nu for rate.

plt.gca().spines['right'].set_visible(False)   ### Make right spine invisible
plt.gca().spines['top'].set_visible(False)     ### Make top spine invisible
plt.gca().spines['left'].set_visible(False)    ### Make top spine invisible
plt.gca().spines['bottom'].set_visible(False)  ### Make top spine invisible

### Output the plot
plt.savefig("plot.pdf")     ### export the plot as this
plt.show()                  ### display the plot in this notebook
```
